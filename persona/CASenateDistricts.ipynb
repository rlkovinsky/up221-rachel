```python
#importing libraries
import pandas as pd
import geopandas as gpd
import plotly.express as px
```

```python
sd = gpd.read_file('CASenateDistricts.geojson')
```

```python
sd.shape
```

```python
sd.head(3)
```

```python
sd.columns = ['Object_ID',
              'GEOID', 
              'SenateDistrict',
              'Senate_District',
              'Senate_District_Area',
              'Shape_Length',
              'Shape_Area',
                'geometry']
```

```python
sd.head(3)
```

```python
sd = sd[['Senate_District',
              'Senate_District_Area',
              'Shape_Length',
              'Shape_Area',
                'geometry']]
sd.head(3)
```

```python
info = gpd.read_file('CPASDInfo.csv')
```

```python
info.head(3)
```

```python
info = info[['Senate_District',
              'Senator',
              'CPA_Communities']]
info.head(3)
```

```python
ca=sd.merge(info,on="Senate_District")
```

```python
ca.sample(3)
```

```python
ca.info()
```

```python
ca = ca[['Senate_District',
         'geometry',
              'Senator',
              'CPA_Communities']]
ca.head(3)
```

```python
# importing 
import matplotlib.pyplot as plt
```

```python
# Making another non-interactive figure with axis turned off and a new title at the top
fig, ax = plt.subplots(figsize = (12,10)) 
ca.plot(column = 'Senate_District',legend=False,
  edgecolor='white',cmap = 'PuBuGn',ax=ax)
ax.axis('off')
plt.title('Testing')
```

```python
# making the interactive map a bit prettier and easier to understand, complete with a legend
c = ca.explore(
     column="Senate_District", # make choropleth based on "neighborhood" column
     popup=True, # show all values in popup (on click)
     tiles="cartodbpositron", # use "CartoDB positron" tiles
     cmap="PuBuGn", # use "PuBuGn" matplotlib colormap
     style_kwds=dict(color="white", weight=".3", fillOpacity=".7"), # use white outline
     zoom_start="7.5"
    )
c
```

```python
c.save('SenateDistricts.html')
```
